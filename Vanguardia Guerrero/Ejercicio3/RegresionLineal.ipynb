# Datasets
Es una colección de datos que se usan para entrenar el modelo. 
Sirven para que los algoritmos "aprendan" a como hacer las predicciones.
Los tipos de datasets más comunes son:
- Datos de Texto
- Datos de Imagen
- Datos de Audio
- Datos de Video
- Datos Numéricos

Primeramente los datos son etiquetados/anotados en orden para que los algoritmos entiendan como debe ser la salida que debe de generar.

## Etiquetado de Datos
Es un proceso de etiquetado de los datos con "Tags" relevantes que permitan a las computadoras entender e interpretar esa información.
Como se etiqueta:
- A través del etiquetado manual
- A través de usar algoritmos de machine learning

![Automated Data Labelling](Images/AutomatedDataLabelling.png)

Para Machine Learning Supervisado, los algoritmos necesitan conocer y entender los patrones de entrada para generar resultados precisos.
- Regresión. Establece la relación entre variables dependientes e independientes.
- Clasificación. Asigna categorías específicas a un conjunto de datos de entrada.

![Supervised Learning Example](Images/SupervisedLearningExample.png)

## Preparación de Datos
Es un prerequisito para el entrenamiento para los algoritmos de la IA y Machine Learning.

### Recolección de Datos. 
El primer paso es la recolección y/o generación de datos relevantes para prepararlos y entrenarlo los algoritmos.
Las diferentes formas son:
- Custom Crowdsourcing
- Recolección privada o interna
- Conjuntos de datos pre-analizados y limpiados
- Recolección automática de datos

De las prácticas recomendadas se puede incluir:
- Entender el problema y las meteas del proyecto IA/ML
- Establecer pipelines de datos o DataOps. s una práctica de gestión de datos automatizada y orientada a procesos. Realiza un seguimiento del ciclo de vida de los datos de principio a fin, lo que proporciona a los usuarios empresariales flujos de datos predecibles.
- Determinar el método de recolección que mejor se adapte a su proyecto
- Evaluación de los datos recopilados para garantizar la calidad
- Recopilación de datos concisos para garantizar la relevancia
- Agregar datos nuevos al conjunto de datos de entrenamiento

# Regresión Lineal

Es un modelo matemático que sirve para aproximar la relación de dependencia entre una variable dependiente Y y m variables independientes Xi y un término aleatorio.

- Predicción entre dos o más variables
- Predicción de un comportamiento

En caso de que no se pueda aplicar este modelo es cuando las variables no tienen una correlación entre si.

- Se genera un hiperplano. Divide 

$Y = B_0+B_1 X_1+...+B_m X_m+E$
